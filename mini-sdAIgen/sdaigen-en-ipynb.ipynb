## 1. File Viewer

```python
!python - <<'PY'
import os, glob, pathlib, folder_paths

roots = [folder_paths.models_dir, "/root/ComfyUI/models", "/kaggle/working/models", "/kaggle/working/ComfyUI/models"]
roots = [r for r in roots if r and os.path.exists(r)]
exts  = ("safetensors","ckpt","pth","pt","gguf","onnx","vae","bin")
limit = int(os.environ.get("SCAN_LIMIT","300"))

print("=== ComfyUI folder_paths ===")
print("base_path:", folder_paths.base_path)
print("models_dir:", folder_paths.models_dir)

seen = set()
for root in roots:
    rp = os.path.realpath(root)
    if rp in seen: 
        continue
    seen.add(rp)
    print(f"\n[ROOT] {rp}")
    sub = ["checkpoints","vae","loras","embeddings","clip","clip_vision","controlnet","upscale_models","ultralytics","sam","diffusers","gguf"]
    present = [s for s in sub if os.path.isdir(os.path.join(rp, s))]
    if present:
        print("subdirs:", ", ".join(present))

    files = []
    for ext in exts:
        files += glob.glob(os.path.join(rp, f"**/*.{ext}"), recursive=True)
    files = sorted(set(files))
    print("files_found:", len(files))
    for p in files[:limit]:
        try:
            sz = pathlib.Path(p).stat().st_size/1e6
        except Exception:
            sz = 0.0
        print(f" - {p} ({sz:.1f} MB)")
    if len(files) > limit:
        print(f" ... +{len(files)-limit} more (raise SCAN_LIMIT to show more)")
PY
```

## 3. Tensor Tester

```python
!python - <<'PY'
import os, glob, pathlib
bad = "qwen_image_edit_fp8_e4m3fn.safetensors"
roots = ["/root/ComfyUI/models","/kaggle/working/models","/kaggle/working/ComfyUI/models"]
hits = []
for r in roots:
    if not os.path.exists(r): continue
    for p in glob.glob(os.path.join(r,"**",bad), recursive=True):
        hits.append(p)
if not hits:
    print("NOT FOUND:", bad); raise SystemExit

for p in hits:
    sz = pathlib.Path(p).stat().st_size
    print("FOUND:", p, f"({sz} bytes)")
print("\nTry opening with safetensors:")
try:
    import safetensors
    from safetensors.torch import load_file
    for p in hits:
        try:
            _ = load_file(p)  # header parse + map
            print("OK:", p)
        except Exception as e:
            print("BROKEN:", p, "|", type(e).__name__, str(e)[:140])
except Exception as e:
    print("safetensors not imported:", e)
PY
```
